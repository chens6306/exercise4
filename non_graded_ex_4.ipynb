```python
import random
import time

# List of all available cards with their attributes
card_deck = [
    {'Name': 'Diablo', 'Health': 100, 'Attack': 90, 'Defense': 60},
    {'Name': 'Medusa', 'Health': 100, 'Attack': 70, 'Defense': 70},
    {'Name': 'Jester', 'Health': 120, 'Attack': 60, 'Defense': 90},
    {'Name': 'Troll', 'Health': 150, 'Attack': 40, 'Defense': 94},
    {'Name': 'Specter', 'Health': 100, 'Attack': 70, 'Defense': 70},
    {'Name': 'Mist', 'Health': 100, 'Attack': 75, 'Defense': 65},
    {'Name': 'Savage', 'Health': 100, 'Attack': 90, 'Defense': 50},
    {'Name': 'Marauder', 'Health': 100, 'Attack': 85, 'Defense': 50},
    {'Name': 'Wimp', 'Health': 110, 'Attack': 40, 'Defense': 85},
    {'Name': 'Sorcerer', 'Health': 100, 'Attack': 70, 'Defense': 55}
]

# Function to show the player's deck
def show_deck(deck):
    for idx, card in enumerate(deck, 1):
        print(f"{idx}. {card['Name']} (Health: {card['Health']}, Attack: {card['Attack']}, Defense: {card['Defense']})")

# Function to calculate and apply attack damage
def attack_phase(attacker_card, defender_card):
    print(f"\n{attacker_card['Name']} attacks {defender_card['Name']}!")
    
    # First, reduce defender's defense
    if defender_card['Defense'] > 0:
        damage_to_defense = min(attacker_card['Attack'], defender_card['Defense'])
        defender_card['Defense'] -= damage_to_defense
        print(f"{defender_card['Name']}'s defense reduced by {damage_to_defense}. New defense: {defender_card['Defense']}")
        
        # If attack is stronger than the remaining defense, reduce health
        remaining_attack = attacker_card['Attack'] - damage_to_defense
        if remaining_attack > 0:
            defender_card['Health'] -= remaining_attack
            print(f"{defender_card['Name']}'s health reduced by {remaining_attack}. New health: {defender_card['Health']}")
    else:
        # If no defense left, reduce health directly
        defender_card['Health'] -= attacker_card['Attack']
        print(f"{defender_card['Name']}'s health reduced by {attacker_card['Attack']}. New health: {defender_card['Health']}")

    # Check if the defender's health drops to 0 or below
    if defender_card['Health'] <= 0:
        print(f"{defender_card['Name']} has been defeated!")
        return True
    return False

# Main game logic
def play_game():
    # Randomly select 5 cards for the player and opponent
    player_deck = random.sample(card_deck, 5)
    opponent_deck = random.sample(card_deck, 5)
    
    # Game loop runs until one deck is empty
    while player_deck and opponent_deck:
        # Show player's deck
        print("\nYour deck:")
        show_deck(player_deck)
        
        # Randomly select an opponent card
        opponent_card = random.choice(opponent_deck)
        print(f"\nOpponent's card: {opponent_card['Name']} (Health: {opponent_card['Health']}, Attack: {opponent_card['Attack']}, Defense: {opponent_card['Defense']})")
        
        # Ask player to select a card
        try:
            while True:
                try:
                    choice = int(input("\nSelect a card to fight with (enter the number): "))
                    if 1 <= choice <= len(player_deck):
                        player_card = player_deck[choice - 1]
                        break
                    else:
                        print("Invalid choice. Try again.")
                except ValueError:
                    print("Please enter a valid number.")
        except KeyboardInterrupt:
            print("\nGame interrupted by user. Exiting...")
            return  # Exit the game loop and terminate the game

        # Player's attack phase
        if attack_phase(player_card, opponent_card):
            opponent_deck.remove(opponent_card)  # Remove defeated opponent card
        
        time.sleep(2)  # Pause for dramatic effect
        
        # Check if opponent has cards left
        if not opponent_deck:
            print("\nCongratulations! You have defeated all the opponent's cards. You win!")
            break
        
        # Opponent's turn to attack
        print("\nOpponent's turn!")
        time.sleep(2)
        
        # Randomly select a player card for opponent to attack
        random_player_card = random.choice(player_deck)
        print(f"Opponent is attacking with {opponent_card['Name']}!")
        
        # Opponent's attack phase
        if attack_phase(opponent_card, random_player_card):
            player_deck.remove(random_player_card)  # Remove defeated player card
        
        time.sleep(2)
        
        # Check if player has cards left
        if not player_deck:
            print("\nAll your cards have been defeated. You lose!")
            break

# Start the game
play_game()
```
